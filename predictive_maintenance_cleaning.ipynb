```python
# delete features that have more than 50% missing values
null_50 = df.isnull().sum()/len(df)*100 > 50
null_50 = null_50[null_50].index
df = df.drop(null_50, axis=1)
df.head(20)
 
```

```python
# lets rename the columns for easy access, replace space with underscore, and convert to lower case, and remove special characters, and remove the word 'engine', and make shorter names
df.columns = df.columns.str.replace(' ', '_').str.lower().str.replace('[-/()#]', '').str.replace('engine', 'eng').str.replace('temperature', 'temp').str.replace('pressure', 'pres').str.replace('vibration', 'vib').str.replace('speed', 'spd').str.replace('temperature', 'temp').str.replace('flow', 'flw')
```

```python
# check how many engines in the data
df['eng_number'].unique()
```

```python
# make a dictionary of engines types and numbers for each type
eng_types = df['eng_type'].unique()
eng_numbers = {}
for eng in eng_types:
    eng_numbers[eng] = df[df['eng_type'] == eng]['eng_number'].unique()
eng_numbers
```

```python
# Change the categorical dat types into suitable type

df['eng_number'] = df['eng_number'].astype('category')
df['eng_type'] = df['eng_type'].astype('category')
df['flight_phase'] = df['flight_phase'].astype('category')
```

### Initial Notes:  
1- We have 3 phases.  
2- From knowledge of the business domain, the critical phases(most stresses) that determine engine life are takeoff and climb.  
3- Also from bussiness domain, each takeoff is considered one running cycle that counts into the engine remaining useful life.  
4- Flight are ordered new to old (last flight of the engine is the first in top).  

```python
# show the null percentage in each column
df.isnull().sum()/len(df)*100
```

```python
# create two cloumns date and time from the flight_datetime, the current format is '25-12-2024  15:05:00 PM'
df['flight_datetime_c'] = pd.to_datetime(df['flight_datetime'], format='%d-%m-%y %H:%M', dayfirst=True)
df['date'] = df['flight_datetime_c'].dt.date
df['time'] = df['flight_datetime_c'].dt.time
df.head(30)
```

```python
#sort the data by engine number and flight datetime
df = df.sort_values(by=['eng_number', 'flight_datetime_c'])
df.head(30)
```

```python
# drop the flight_datetime column
df = df.drop('flight_datetime', axis=1)
```

```python
# for each engine lets assign a flight cycle for each row, flight cycle start with take off and end with cruise
# flight cycle will increase by 1 for each take off from oldest to newest date for each engine
# first split daata by engine into a list of dataframes 
# then from oldest to newsest date for each engine assign a flight cycle for each row that increaes by 1 for each take off
# then combine the dataframes into one dataframe
df['flight_cycle'] = 0
df_list = [df[df['eng_number'] == i] for i in df['eng_number'].unique()]
for i in range(len(df_list)):
    df_list[i] = df_list[i].sort_values(by='flight_datetime_c')
    flight_cycle = 0
    for j in range(len(df_list[i])):
        if df_list[i].iloc[j]['flight_phase'] == 'TAKEOFF':
            flight_cycle += 1
        df_list[i].iloc[j, df_list[i].columns.get_loc('flight_cycle')] = flight_cycle
```

```python
# combine the dataframes
df = pd.concat(df_list)
```

```python
df.head(20)
```

```python
# add a column for remaining useful life (RUL) for each engine
# RUL is the number of flight cycles remaining for the engine before it fails
# RUL = max(flight_cycle for this engine) - flight_cycle
# store the max flight cycle for each engine in a dictionary
max_flight_cycle = {}
for i in df['eng_number'].unique():
    max_flight_cycle[i] = df[df['eng_number'] == i]['flight_cycle'].max()
df['RUL'] = 0
for i in range(len(df)):
    df.iloc[i, df.columns.get_loc('RUL')] = max_flight_cycle[df.iloc[i]['eng_number']] - df.iloc[i]['flight_cycle']
```

```python
# lets check the data
df.head(20)
```

```python
max_flight_cycle
```

```python
# lets add  max_flight_cycle dict to the engine life dict in days
engine_life = {}
for i in df['eng_number'].unique():
    engine_life[i] = (df[df['eng_number'] == i]['date'].max() - df[df['eng_number'] == i]['date'].min()).days

# Add max_flight_cycle to engine_life
for i in engine_life.keys():
    engine_life[i] = {'Life in Days': engine_life[i], 'Life in FC': max_flight_cycle[i]}
engine_life
```

```python
# save the cleaned data to a csv file
df.to_csv('engines_data_cleaned.csv', index=False)
```
